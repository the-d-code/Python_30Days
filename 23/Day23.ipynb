```python
# read data
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np # linear algebra
import plotly.express as px
import plotly.graph_objects as go # or plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime

df = pd.read_csv("NFLX.csv")
df
```

```python
# top data
df.head()
```

```python
# bottom data
df.tail()
```

```python
sns.set(rc={'figure.figsize':(10, 5)})
```

```python
# change index

# correct date formate
df['Date'] = pd.to_datetime(df['Date'])
# set index
df = df.set_index('Date')
# sort index
df = df.sort_index()
df
```

```python
# volume of stock traded

sns.lineplot(x=df.index, y=df['Volume'])
plt.title("Volume of Stock Traded")
plt.show()
```

```python
# volume of stock traded

sns.lineplot(x=df.index, y=df['Volume'], label="Volume", color='blue')
plt.title("Volume of Stock Traded")

plt.show()
```

```python
# netflix stock price : high, open, low, close

sns.lineplot(x=df.index, y=df['High'], label="High")
sns.lineplot(x=df.index, y=df['Open'], label="Open")
sns.lineplot(x=df.index, y=df['Low'], label="Low")
sns.lineplot(x=df.index, y=df['Close'], label="Close")
plt.title("Stock Price")
plt.ylabel("Price")
plt.show()
```

```python
# netflix stock price : Day, Month, Year wise chart

fig, (ax1, ax2, ax3) = plt.subplots(3,figsize=(15, 10))

df.groupby(df.index.day).mean().plot(y = 'Volume', ax=ax1,xlabel="Day", ylabel="Volume", title="Day wise Volume")
df.groupby(df.index.month).mean().plot(y = 'Volume', ax=ax2,xlabel="Month", ylabel="Volume", title="Month wise Volume")
df.groupby(df.index.year).mean().plot(y = 'Volume', ax=ax3,xlabel="Year", ylabel="Volume", title="Year wise Volume")

plt.show()
```

```python
# top 5 dates with highest stock price

a = df.sort_values(by='High', ascending=False).head(5)
a['High']
```

```python
# top 5 dates with lowest stock price

a = df.sort_values(by='Low', ascending=True).head(5)
a['Low']
```

```python
# plot graph

fig,axes = plt.subplots(nrows=1, ncols=2, sharex=True, figsize=(15, 5))
fig.suptitle('High and Low Stock Price per period of time', fontsize=16)

sns.lineplot(ax=axes[0], x=df.index, y=df['High'], label="High", color='green')
sns.lineplot(ax=axes[1], x=df.index, y=df['Low'], label="Low", color='red')

plt.show()
```
