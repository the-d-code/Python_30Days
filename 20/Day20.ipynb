```python
# Plotly
x = [1, 2, 3, 4, 5]
y = [1, 2, 4, 8, 16]
```

```python
# matplotlib
import matplotlib.pyplot as plt
plt.plot(x, y)
```

```python
# seaborn
import seaborn as sns

x = [1, 2, 3, 4, 5]
y = [1, 2, 4, 8, 16]


fig = sns.lineplot(x=x, y=y)
plt.show()
```

```python
# plotly
import plotly.express as ps

x = [1, 2, 3, 4, 5]
y = [1, 2, 4, 8, 16]

fig = ps.line(x=x, y=y, title="Plotly Line Chart")
fig.show()
```

```python
import plotly.graph_objects as go

x = [1, 2, 3, 4, 5]
y = [1, 2, 4, 8, 16]

fig = go.Figure(data=go.Scatter(x=x, y=y))
fig.show()
```

# adding titles

```python
# add title

fig = go.Figure(data=go.Scatter(x=x, y=y))

fig.update_layout(title="Plotly Line Chart", xaxis_title="this is x", yaxis_title="this is y")
fig.show()
```

```python
x1, y1 = [1, 2, 3, 4, 5], [1, 2, 4, 8, 16]
x2, y2 = [1, 2, 3, 4, 5], [1, 3, 9, 12, 15]
x3, y3 = [1, 2, 3, 4, 5], [1, 4, 16, 11, 64]

fig = go.Figure()
fig.add_trace(go.Scatter(x=x1, y=y1, name="X1Y1"))
fig.add_trace(go.Scatter(x=x2, y=y2, name="X2Y2"))
fig.add_trace(go.Scatter(x=x3, y=y3, name="X3Y3"))
fig.update_layout(title="Plotly Line Chart", xaxis_title="this is x", yaxis_title="this is y")
fig.show()
```

# buble chart

```python
import plotly.graph_objects as go
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 4, 8, 16]

fig = go.Figure(data=go.Scatter(x=x, y=y, mode="markers", 
                                marker=dict(size=[20, 40, 60, 80, 100],
                                            
                                            color=[0, 1, 2, 3, 4],
                                            colorscale="Viridis")))
fig.show()
```
